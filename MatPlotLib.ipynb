```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#this is a magic command in Jupyter Notebook! by this command we will be displaying plots inline not in a new window!
```

```python
import numpy as np
x = np.arange(0,5*np.pi,0.1)
y1 = np.sin(x)
plt.plot(x,y1)
print(x)
```

```python
y2 = np.cos(x)
plt.plot(y1)
plt.plot(y2)
plt.xlabel('x-axis')
plt.ylabel('y_axis')
plt.title('trigonometric functions in x')
plt.legend(['sine','cosine'])
```

```python
plt.plot(x,y1,'r--',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'yx',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'r*',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'r*',x,y2,'bo')
plt.legend(['sin','cos'])
```
