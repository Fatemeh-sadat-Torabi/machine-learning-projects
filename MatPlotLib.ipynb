```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#this is a magic command in Jupyter Notebook! by this command we will be displaying plots inline not in a new window!
```

```python
import numpy as np
x = np.arange(0,5*np.pi,0.1)
y1 = np.sin(x)
plt.plot(x,y1)
print(x)
```

```python
y2 = np.cos(x)
plt.plot(y1)
plt.plot(y2)
plt.xlabel('x-axis')
plt.ylabel('y_axis')
plt.title('trigonometric functions in x')
plt.legend(['sine','cosine'])
```

```python
plt.plot(x,y1,'r--',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'yx',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'r*',x,y2,'k:')
plt.legend(['sin','cos'])
```

```python
plt.plot(x,y1,'r*',x,y2,'bo')
plt.legend(['sin','cos'])
```

# Subplot:

```python
x = np.arange(0,3*np.pi,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
# setup subplot grid with height 2 width 1, and set first subplot as active!
plt.subplot(2,1,1)
# make first plot:
plt.plot(x,y1)
plt.title('sin')
# set second subplot as active, and make second plot:
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('cos')
# show the figure:
plt.show()
```

```python
x = np.linspace(-2,2,40)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(x,x**3)
plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.show()
```

How to save a figure in python:

```python
x = np.linspace(-2,2,400) # 400 numbers in range -2 , 2 with equal space (linear space)
y = x**2
plt.plot(x,y,'r--')
plt.xlabel('$x$')
plt.ylabel('$y=x^2$')
plt.title('square function')
plt.savefig('x2.png')
plt.show()
```
